# Make prediction with Decision Tree Regression

## Data preprocessing

### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import the data

```python
data = pd.read_csv("Position_Salaries.csv")

# Turn data into inputs and outputs
x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
print(x.shape)
```

```python
# Visualizing data with seaborn library
sns.set_theme(style="darkgrid")
sns.scatterplot(data=data, x='Level', y='Salary', style='Position', s=100)
plt.show()
```

## Training the model

### Build the Decision Tree Regression model

```python
from sklearn.tree import DecisionTreeRegressor
dtr_model = DecisionTreeRegressor(random_state=0)
dtr_model.fit(x, y)
```

## Model prediction

```python
y_pred = dtr_model.predict(x)
y_pred
```

```python
dtr_model.predict([[6.5]])
```

## Model performance visualizing

```python
x_grid = np.arange(x.min(), x.max()+0.1, 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
x_grid
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=x[:, 0], y=y, mode='markers', name='ground truth', hovertemplate="Level:%{x}<br>Salary:%{y}", marker=dict(color="blue")))
fig.add_trace(go.Scatter(x=x_grid[:, 0], y=dtr_model.predict(x_grid), name='predicted by Decison tree', hovertemplate="Level: %{x}<br>Salary: %{y}<extra></extra>"))
fig.update_layout(title="Decesion Tree Regressoin ",xaxis_title="Level", yaxis_title="Salary")
fig.show()
```
