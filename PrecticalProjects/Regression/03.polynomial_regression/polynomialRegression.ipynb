#Polynomial Regression

##import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##importing the dataset

```python
data = pd.read_csv("Position_Salaries.csv")
```

##slicing the data

```python
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
x
```

##spliting the data into test and training

```python
# from sklearn.model_selection import train_test_split

# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
#our data is to small so we skip from this part
```

##training the linear regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression

lr_modeling = LinearRegression()
lr_modeling.fit(x,y)
```

##trining the model with polinomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
pol_regression = PolynomialFeatures(degree = 5)
x_poly = pol_regression.fit_transform(x)
lr_modeling_2 = LinearRegression()
lr_modeling_2.fit(x_poly,y)
```

##visualising the linear regression result

```python
plt.scatter(x, y, color = "red")
plt.plot(x, lr_modeling.predict(x), color = "blue")
plt.title('predict with linear regression')
plt.xlabel("year of experience")
plt.ylabel('salary')
plt.show()
```

##visualising with polynomial regression result

```python
plt.scatter(x, y, color = "red")
plt.plot(x, lr_modeling_2.predict(x_poly ), color = "blue")
plt.title("predict with polynomial regression")
plt.xlabel('the position level')
plt.ylabel('salary')
plt.show()
```

##visualising the polynomial regression with higher resolution

```python
x_grid = np.arange(min(x),max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(x, y, color = "red")
plt.plot(x_grid, lr_modeling_2.predict(pol_regression.fit_transform(x_grid) ), color = "blue")
plt.title("predict with polynomial regression")
plt.xlabel('the position level')
plt.ylabel('salary')
plt.show()
```

##predicting new result with linear regression

```python
lr_modeling.predict([[6.5]])
```

##predicting new result with polynomial regression

```python
lr_modeling_2.predict(pol_regression.fit_transform([[6.5]]))
pol_regression.fit_transform([[6.5]])
```
